3Ô∏è‚É£ AN√ÅLISES ESTAT√çSTICAS üìä

Calculo de frequ√™ncias (absoluta e relativa)

#### Calculo das frequ√™ncias absoluta, relativa e percentual de uma vari√°vel num√©rica.

Par√¢metros: serie (pd.Series) - Coluna do DataFrame contendo os dados categorizados.
Retorna: DataFrame com frequ√™ncias absoluta, relativa e percentual.

```
freq_absoluta = Counter(idades)
```

```
freq_absoluta
```

```
freq_absoluta = pd.DataFrame.from_dict(freq_absoluta, orient='index')
```

```
freq_absoluta
```

```
#Ordenar por ordem crescente de idade
freq_absoluta = freq_absoluta.sort_index(ascending=True)
freq_absoluta
```

```
freq_relativa = freq_absoluta / freq_absoluta.sum()
freq_relativa
```

```
freq_relativa_perc = freq_relativa * 100
freq_relativa_perc
```

```
#Contando a frequ√™ncia de cada idade na base de dados enem_sorocaba
frequencias_idade = enem_sorocaba['IDADE'].value_counts()
```

```
#Mostrando a contagem das idades
frequencias_idade
```

```
# üìå Exibindo as 10 idades mais frequentes
frequencias_idade.head(10)
```

4Ô∏è‚É£ VISUALIZA√á√ÉO DE DADOS (GR√ÅFICOS E HISTOGRAMAS)


**Matplotlib**

```
# Importa a biblioteca para visualiza√ß√£o de gr√°ficos
import matplotlib.pyplot as plt
```

### Geramos um histograma com estiliza√ß√£o aprimorada.

    Par√¢metros:
    - serie (pd.Series) ‚Üí Dados num√©ricos a serem visualizados.
    - bins (int) ‚Üí N√∫mero de intervalos no histograma (padr√£o = 10).
    - titulo (str) ‚Üí T√≠tulo do gr√°fico.
  

```
# Foi gerado um histograma b√°sico para visualizar a distribui√ß√£o de idades
enem_sorocaba.IDADE.hist(bins=9) #Se refere ao n√∫mero de colunas amostradas no gr√°fico
plt.title('Distribui√ß√£o de Idades - Enem SP 2019 - Sorocaba')
plt.show()
```

```
# Cria√ß√£o de um histograma com mais ajustes visuais
plt.hist(enem_sorocaba.IDADE, 8, rwidth=1)
plt.title('Distribui√ß√£o de Idades - Enem SP 2019 - Sorocaba')
plt.show()
```

```
# Histograma com estiliza√ß√£o e importa√ß√£o de cor personalizada
from matplotlib import colors
plt.hist(enem_sorocaba.IDADE, 10, rwidth=1, color='Red', density = True, )
plt.title('Distribui√ß√£o de Idades - Enem SP 2019 - Sorocaba')
plt.xlabel ('Idade')
plt.ylabel ('Frequ√™ncia')
plt.show()
```
